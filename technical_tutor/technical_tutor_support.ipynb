## Technical Tutor Chat Bot

Explain technical information for given user prompt.

```python
import os
import json
from dotenv import load_dotenv
import ollama
```

```python
load_dotenv(override=True)

MODEL= 'qwen2.5:0.5b'
```

```python
technical_tutor_system_prompt = """
Answer user's technical questions and give him detail explanation
"""

user_prompt = """
Explain to me what is AI engineering?
"""
```

```python
def technical_tutor_chat(question):
    response = ollama.chat(
        model=MODEL,
        messages=[
            {"role": "system", "content": technical_tutor_system_prompt},
            {"role": "user", "content": question}
        ]
    )

    result = response['message']['content']
    return result
```

```python
technical_tutor_chat(user_prompt)
```
